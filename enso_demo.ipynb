```python
# Find the market of interest
markets_df = get_markets_list()
markets_df[markets_df.market_name.str.contains("Los")]
```

```python
# Find the commodities sold in that market
slug_id = "2306"  # Los Angeles fruits market
commodities_list = get_commodities_list(slug_id = slug_id)
commodities_list.sort()
pprint(commodities_list)
```

```python
# Get specific commodity's prices
commodity = "Mangoes"
prices_df = get_prices(commodity=commodity, slug_id=slug_id, start_year=2012, end_year=2018).dropna(subset=['price'])
prices_df['date'] = pd.to_datetime(prices_df['date'], format="%Y-%m-%d")
prices_df.head()
```

```python
print("Varieties")
varieties = prices_df.variety.value_counts()
pprint(varieties)
print("--------------------------")
print("Packages")
packages = prices_df.package.value_counts()
pprint(packages)
print("--------------------------")
print("Item sizes")
item_sizes = prices_df.item_size.value_counts()
pprint(item_sizes)
print("--------------------------")
print("Origins")
origin = prices_df.origin.value_counts()
pprint(origin)
```
